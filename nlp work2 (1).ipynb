```python
data=pd.read_csv("News Sentiment.csv")
```

```python
data=data.drop(["IDLink","Headline","Source","Topic","PublishDate","Facebook","GooglePlus","LinkedIn","SentimentHeadline"],axis=1)
```

```python
import gensim
```

```python
import string
```

```python
def remove_punctuation(text):
    a=''.join([i.lower() for i in text if i not in string.punctuation])
    return a
```

```python
data["new title"]=data["Title"].apply(lambda i:remove_punctuation(i))
```

```python
data=data.drop(["SentimentTitle"],axis=1)
```

```python
import re
```

```python
def remove_num(text):
    text=re.sub('[0-9]+', '',text )
    return text
```

```python
data['new title']=data['new title'].apply(lambda i:remove_num(i))
```

```python
data
```

```python
from nltk.tokenize import word_tokenize
# from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
token=[]
lemme= WordNetLemmatizer()
stop_words=stopwords.words("english")
for i in data["new title"]:
    word=word_tokenize(i)
    lemmatize=[lemme.lemmatize(i) for i in word if i not in stop_words]
    token.append(' '.join(lemmatize))
    
```

```python
data["simple title"]=data["Title"].apply(lambda i:gensim.utils.simple_preprocess(i))
```

```python
# tagging
tagged_document=[]
for i,v in enumerate(data["simple title"]):
    tagged_document.append(gensim.models.doc2vec.TaggedDocument(v,[i]))
```

```python
# tagged_document
```

```python
model=gensim.models.Doc2Vec(tagged_document,window=5,vector_size=100,min_count=2)
```

```python
print(model)
```

```python
model.save("doc2vec1.model")
```

```python
a=model.infer_vector("Listening,  Reading and Writing (L-S-R-W) are the four skills of language learning".split())
```
